# Using Neural Networks to Predict if an MRI scan of a Brain contains a tumor and detect where it is located. 
 In this repository we will be working towards proving that not only can Neural Network can accurately predict if cancer is in an MRI scan, but make a guess at where its located in the image. 

 This presentation was made for the GA Medical Conference, where it will be presented to doctors, hospital staff, and hospital administration. This model is intended to assist doctor's in diagnosis and save expenditures for hospitals. 

---

## How is success measured?
> I will be using a functional classification CNN model using TensorFlow and Keras. The data the model will be trained upon was gathered from [Kaggle](https://www.kaggle.com/datasets/masoudnickparvar/brain-tumor-mri-dataset?select=Testing) and contains 7023 images of MRI scans of brains in 4 categories: 'Glioma', 'Meningioma', 'No Tumor', and 'Pituitary'. The goal for the model is to achieve the highest accuracy, with as little loss as possible. When making anything for the medical field, people's lives can be at stake, so it's important to aim for an accuracy that's not just above baseline (0.28) but nearing as close to 100% as possible. 

> Grad-CAM will be utilized post-modeling to create visualizations of the model's weights. The hope is that after learning the patterns of cancerous tumors, it will "see" a tumor in the image and it will be a heavy weight in the feature map. So when we receive the heatmap overlayed on the scan, it will show where it believes a tumor is located.
